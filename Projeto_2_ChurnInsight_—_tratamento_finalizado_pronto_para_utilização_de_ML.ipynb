```python
# Importa as bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# URL do arquivo CSV no repositório GitHub (conteúdo raw)
github_url = 'https://raw.githubusercontent.com/ElenirOliveira/churn-insight-api/main/WA_Fn-UseC_-Telco-Customer-Churn.csv'

try:
    dados_github = pd.read_csv(github_url)
    print("Dados carregados com sucesso do GitHub!")
except Exception as e:
    print(f"Erro ao carregar dados do GitHub: {e}")
```

```python
dados_github.head()
dados_github.shape
dados_github.info()
dados_github.describe(include='all')
```

```python
columns_traducao = {
    'customerID': 'ID_Cliente',
    'gender': 'Genero',
    'SeniorCitizen': 'Cidadao_Senior',
    'Partner': 'Parceiro',
    'Dependents': 'Dependentes',
    'tenure': 'Permanencia_Meses',
    'PhoneService': 'Servico_Telefone',
    'MultipleLines': 'Multiplas_Linhas',
    'InternetService': 'Servico_Internet',
    'OnlineSecurity': 'Seguranca_Online',
    'OnlineBackup': 'Backup_Online',
    'DeviceProtection': 'Protecao_Dispositivo',
    'TechSupport': 'Suporte_Tecnico',
    'StreamingTV': 'TV_Streaming',
    'StreamingMovies': 'Filmes_Streaming',
    'Contract': 'Contrato',
    'PaperlessBilling': 'Fatura_Sem_Papel',
    'PaymentMethod': 'Metodo_Pagamento',
    'MonthlyCharges': 'Cobrancas_Mensais',
    'TotalCharges': 'Cobrancas_Totais',
    'Churn': 'Evasao'
}

dados_github = dados_github.rename(columns=columns_traducao)
```

```python
translation_map = {
    'Female': 'Feminino',
    'Male': 'Masculino',
    'Yes': 'Sim',
    'No': 'Nao',
    'No phone service': 'Sem_servico_telefone',
    'DSL': 'DSL',
    'Fiber optic': 'Fibra_otica',
    'No internet service': 'Sem_servico_internet',
    'Month-to-month': 'Mensal',
    'One year': 'Um_ano',
    'Two year': 'Dois_anos',
    'Electronic check': 'Cheque_eletronico',
    'Mailed check': 'Cheque_enviado',
    'Bank transfer (automatic)': 'Transferencia_bancaria',
    'Credit card (automatic)': 'Cartao_credito'
}

for col in dados_github.select_dtypes(include='object').columns:
    if col != 'ID_Cliente':
        dados_github[col] = dados_github[col].replace(translation_map)
```

```python
for col in dados_github.select_dtypes(include='object').columns:
    dados_github[col] = dados_github[col].str.strip().str.lower()
```

```python
dados_github.columns = (
    dados_github.columns
    .str.normalize('NFKD')
    .str.encode('ascii', errors='ignore')
    .str.decode('utf-8')
    .str.lower()
)
```

```python
colunas_numericas = [
    'permanencia_meses',
    'cobrancas_mensais',
    'cobrancas_totais'
]

dados_github[colunas_numericas] = dados_github[colunas_numericas].apply(
    pd.to_numeric, errors='coerce'
)
```

```python
dados_github['cobrancas_totais'] = dados_github['cobrancas_totais'].fillna(0)
dados_github.isnull().sum()
```

```python
dados_github = dados_github.drop_duplicates(subset='id_cliente')
```

```python
dados_github['evasao'] = dados_github['evasao'].map({'sim': 1, 'nao': 0})
```

```python
dados_github['ticket_medio'] = (
    dados_github['cobrancas_totais'] / (dados_github['permanencia_meses'] + 1)
)

dados_github['cliente_novo'] = dados_github['permanencia_meses'].apply(
    lambda x: 1 if x <= 6 else 0
)
```

```python
X = dados_github.drop(columns=['evasao', 'id_cliente'])
y = dados_github['evasao']
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

categoricas = X.select_dtypes(include='object').columns
numericas = X.select_dtypes(exclude='object').columns

preprocessador = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numericas),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categoricas)
    ]
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

pipeline = Pipeline(steps=[
    ('preprocessamento', preprocessador),
    ('modelo', LogisticRegression(max_iter=1000))
])
```
